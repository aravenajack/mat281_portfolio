EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import numpy as np
import os
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df
```

```python
def nan_percol(df):
    columnasdf=list(df)
    nan_percols=[0 for i in columnasdf]
    j=0
    for i in columnasdf:
        nan_percols[j]=len(df[ df[i].isnull() ])
        j=j+1
    return nan_percols

nan_percol(df)
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

__Respuesta:__ Para eliminar los valores Nan debemos identificar cada fila que tenga al menos un nan y discriminar que hacer con la fila, eliminarla o rellenar. En este caso, con la función 'nan_percol()' se chequeamos cada columna para determinar cuantos elementos nan tenemos por columna. En base a la respuesta que da la función, tenemos que hay 3 valores Nan en la última columna. Por tanto para eliminar los valores Nan, simplemente aplicamos un filtro según la última columna.

```python
print('Antes teniamos', len(df) ,'datos' )
df1=df[ df['price'].notnull() ]
print('Ahora tenemos', len(df1), 'datos')
```

Efectivamente, se eliminan 3 filas. ¿Cuales fueron eliminadas?

```python
df[ df['price'].isnull() ]
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
grupo = df1.groupby(['company'])
df_maxprice_percomp = grupo.agg({'price':[np.max]}).reset_index()
maxp = df_maxprice_percomp['price']['amax'].max()
mask = df_maxprice_percomp['price']['amax']==maxp
df_maxprice_percomp[mask]
```

Para determinar el nombre de la compañia de automóviles más cara, pensamos en aquella compañia que tenga el modelo de automóvil más caro. Para encontrarlo aplicamos un groupby por compañía y luego identificamos el más caro por cada compañía. A este dataframe le aplicamos una máscara para identificar el máximo. Con ello obtenemos que la marca Mercedes Benz es la marca más cara.

3. Imprimir todos los detalles de Toyota Cars

```python
companys=df1['company'].unique()
```

```python
df1[ df1['company']=='toyota' ]
```

4. Cuente el total de automóviles por compañía

```python
df_numpercomp=grupo['price'].count().reset_index()
df_numpercomp.columns = ['company', 'n° cars']
df_numpercomp
```

5. Encuentra el coche con el precio más alto por compañía

```python
expncar_percomp=pd.DataFrame()
j=0
for i in companys:
    cars_comp_i = df1[ df1['company']==i ]
    expncar_percomp = pd.concat( [ expncar_percomp , cars_comp_i[ cars_comp_i['price']==cars_comp_i['price'].max() ]  ] )
    j+1
expncar_percomp
```

Notemos que tenemos dos autos alfa-romero con el mismo precio.

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
df_mileaverage_percomp = grupo.agg({'average-mileage':[np.mean]}).reset_index()
df_mileaverage_percomp
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
df1.sort_values(by=['price'], ascending=False)
```
