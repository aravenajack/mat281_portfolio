```python

import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import math


sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df
```

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
palette = sns.color_palette("hls", 2)

fig = sns.lineplot(
    x='month_number',
    y='total_profit',
    data=df,
    palette=palette)
```

```python


fig = plt.figure(figsize=(10, 5)) # ventana
plt.scatter(
    df['month_number'], # eje x
    df[ 'total_profit' ], # eje y
    c="g", # color
    alpha=0.7, # intensidad color
    label="Total Profit" # etiquetdo fijando posicion,
)
plt.xlabel('Month')
plt.ylabel('Total Profit')
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
nombresdf=list(df)
nombres_de_interes=nombresdf[1:7]
nombres_de_interes
```

```python
fig = plt.figure(figsize=(10, 10)) 

plt.plot(
    df['month_number'], # eje x
    df['facecream'], # eje y
    label='facecream', # etiquetado
    color="b", # color
    linewidth=3 # tamano de la curva
)

plt.plot(
    df['month_number'], # eje x
    df['facewash'], # eje y
    label='facewash', # etiquetado
    color="r", # color
    linewidth=6 # tamano de la curva
)

plt.plot(
    df['month_number'], # eje x
    df['toothpaste'], # eje y
    label='toothpaste', # etiquetado
    color="g", # color
    linewidth=3 # tamano de la curva
)

plt.plot(
    df['month_number'], # eje x
    df['bathingsoap'], # eje y
    label='bathingsoap', # etiquetado
    color="m", # color
    linewidth=3 # tamano de la curva
)

plt.plot(
    df['month_number'], # eje x
    df['shampoo'], # eje y
    label='shampoo', # etiquetado
    color="y", # color
    linewidth=3 # tamano de la curva
)

plt.plot(
    df['month_number'], # eje x
    df['moisturizer'], # eje y
    label='moisturizer', # etiquetado
    color="c", # color
    linewidth=3 # tamano de la curva
)

plt.legend()
plt.ylabel('Unidades Vendidas')
plt.xlabel(' Mes ')
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
Data = df[ ['month_number', 'facecream' , 'facewash']]
```

```python
fig, ax = plt.subplots()
index = np.arange(12)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, df['facecream'], bar_width,
alpha=opacity,
color='b',
label='Facecream')

rects2 = plt.bar(index + bar_width, df['facewash'] , bar_width,
alpha=opacity,
color='g',
label='Facewash')

plt.xlabel('Meses')
plt.ylabel('Unidades Vendidas')
plt.title('Unidades vendidas por mes')
plt.xticks(index + bar_width, df['month_number'])
plt.legend()

plt.tight_layout()
plt.show()
```
